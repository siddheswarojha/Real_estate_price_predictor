```python
import pandas as pd
import numpy as nmp
import matplotlib.pyplot as plt
from sklearn import linear_model
import joblib as jb
from sklearn.model_selection import train_test_split
```

```python
df1 = pd.read_csv("Bengaluru_House_Data.csv")
df1
```

```python
df2=df1.drop(["area_type","availability","society"],axis="columns")
df2
```

```python
df2.isnull().sum()
```

```python
median = df2.balcony.median()
median
```

```python
df2.balcony= df2.balcony.fillna(median)
df2.head()
```

```python
df2.isnull().sum()
```

```python
mdn = df2.bath.median()
mdn
```

```python
df2.bath=df2.bath.fillna(mdn)
df2.isnull().sum()
```

```python
df3 = df2.dropna()
df3.isnull().sum()
```

```python
df3
```

```python
df3['size'].unique()
```

```python
df3['bhk']=df3['size'].apply(lambda x: int(x.split(' ')[0]))
df3
```

```python
df4= df3.drop(['size'],axis='columns')
df4
```

```python
df4['total_sqft'].unique()
```

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
df4[~df4['total_sqft'].apply(is_float)].head(10)
```

```python
# def conv_sqmt_to_sqft(x):
#     x=x*10.764
#     return x
    
```

```python
# def conv_perch_to_sqft(x):
#     x=x*272.3
#     return x
```

```python
def conv_range_to_float(x):
    token=x.split('-')
    if len(token)==2:
        return (float(token[0])+float(token[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
df5 = df4.copy();
df5['total_sqft']=df5['total_sqft'].apply(conv_range_to_float)
df5.head()
```

```python
#location based synchronization for predicting the model of prices

location_stats = df5.groupby('location')['location'].agg('count').sort_values(ascending=False)
location_stats
```

```python
location_stats_less_than_ten = location_stats[location_stats<10]
```

```python
df5.location = df5.location.apply(lambda x: 'other' if x in location_stats_less_than_ten else x)
```

```python
len(df5.location.unique())
```

```python
df5.head(100)
```

```python
newDF = pd.get_dummies(df5.location)
newDF
```

```python
mergedDF = pd.concat([df5,newDF],axis ='columns')
mergedDF
```

```python
finalDF = mergedDF.drop(['location','other'],axis = 'columns')
finalDF
```

```python
finalDF.dropna()
```

```python
finalDF.isnull().sum()
```

```python
mid = finalDF.total_sqft.median()
mid
```

```python
finalDF.total_sqft = finalDF.total_sqft.fillna(mid)
finalDF
```

```python
X=finalDF.drop(['price'],axis='columns')
X
```

```python
y= df5.price
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size =0.01)
```

```python
price_prediction_model = linear_model.LinearRegression()
```

```python
price_prediction_model.fit(X_train,y_train)
```

```python
price_prediction_model.predict(X_test)
```

```python
price_prediction_model.score(X_test,y_test)
```

```python
jb.dump(price_prediction_model,"bangalore_price_prediction_model_for_real_estate")
```
